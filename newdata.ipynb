### which columns to encode

```python
to_encode = ['school', 'sex', 'address', 'famsize', 'Pstatus',
       'Mjob', 'Fjob', 'reason', 'guardian',
        'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic','result']
```

```python
df['result']=df['G1']+df['G2']+df['G3']
df['result'] = df['result'].apply(lambda x: 'fail' if x < 20 else 'pass')
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()

# Encode each categorical column
for col in to_encode:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

print(df_encoded.head())
```

```python
df_encoded.shape
```

### Dropping the G1, G2 and G3 columns as we have created an extra column 'result' by adding G1 , G2 and G3

```python
df_encoded.drop('G1',axis=1,inplace=True)
```

```python
df_encoded.drop('G2',axis=1,inplace=True)
```

```python
df_encoded.drop('G3',axis=1,inplace=True)
```

```python
df_encoded
```

## Selecting the 15 best features

```python
X = df_encoded.drop(columns=['result'])
y=df_encoded['result']
```

```python
from sklearn.feature_selection import SelectKBest, chi2
selector = SelectKBest(score_func=chi2, k=15)
X_new = selector.fit_transform(X, y)
```

```python
selected_cols = X.columns[selector.get_support()].tolist()

print("Best 15 features:", selected_cols)
```
